```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
import seaborn as sns
```

```python
file_path = '../data/graphs.xlsx' 
df = pd.read_excel(file_path)
```

# Comparison of Actual, CCD, and Random Forest Outputs

```python
required_columns = ['actual', 'ccd', 'random_forest']

plt.figure(figsize=(10, 6))
plt.plot(df.index, df['actual'], label='Actual', marker='o')
plt.plot(df.index, df['ccd'], label='CCD', marker='o')
plt.plot(df.index, df['random_forest'], label='RF', marker='o')

plt.xlabel('Iteration (Row Index)')
plt.ylabel('Values')
plt.title('Comparison of Actual, CCD, and Random Forest Outputs')
plt.legend()

plt.xticks(ticks=range(df.index.min(), df.index.max() + 1, 1))

plt.show()
```

# Scatter plot

```python
scatter_matrix(df, figsize=(12, 12), diagonal='kde')
plt.show()
```

# Corelation map

```python
corr = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.show()
```

# Pairwise Regression

```python
sns.pairplot(df, kind='reg')
plt.show()
```
